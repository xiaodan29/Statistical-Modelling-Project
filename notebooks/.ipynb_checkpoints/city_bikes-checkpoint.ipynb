# CityBikes

Send a request to CityBikes for the city of your choice. 

```python
import requests
import pandas as pd
```

```python
# Step 1: Explore the API structure and query the API
# select city = Vancouver(id= "mobibikes" for this website) to explore first
networks_url = 'https://api.citybik.es/v2/networks/mobibikes'
vancouver = requests.get(networks_url).json()
vancouver
```

Parse through the response to get the details you want for the bike stations in that city (latitude, longitude, number of bikes). Put your parsed results into a DataFrame.

```python
stations = vancouver['network']['stations']
data = []

for station in stations:
    latitude = station['latitude']
    longitude = station['longitude']
    num_bikes = station['free_bikes']
    uniqueId = station['id']
    data.append({'latitude': latitude, 'longitude': longitude, 'num_bikes': num_bikes, 'uniqueId': uniqueId})

van_citybike = pd.DataFrame(data, columns=['latitude', 'longitude', 'num_bikes', 'uniqueId'])

van_citybike.head(15)
```
