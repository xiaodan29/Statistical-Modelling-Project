Join the data from Part 1 with the data from Part 2 to create a new dataframe.

```python
import pandas as pd

# Load the van_citybike, foursquare, and yelp dataframes from the CSV files
van_citybike = pd.read_csv('van_citybike.csv')
foursquare = pd.read_csv('van_data.csv')
yelp = pd.read_csv('yelp_for_citybike.csv')

```

```python
# Merge the van_citybike and foursquare dataframes using a left join on the latitude and longitude columns
merged_data = pd.merge(van_citybike, foursquare, on=['latitude', 'longitude', 'num_bikes', 'uniqueId'], how='left')
```

```python
# Merge the merged_data and yelp dataframes using a left join on the latitude and longitude columns
citybike_combin = pd.merge(merged_data, yelp, on=['latitude', 'longitude'], how='left')

```

```python
# Save the merged dataframe to a CSV file
citybike_combin.to_csv('citybike_combin.csv', index=False)
```

```python
# Check for NaNs in the new merged dataframe-citybike_combin
if citybike_combin.isna().any().any():
    print('The merged dataframe contains NaN values.')
else:
    print('The merged dataframe does not contain any NaN values.')
```

Provide a visualization that you used as part of your EDA process. Explain the initial pattern or relationship you discoved through this visualization. 

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Before do the visualization, I already check the NaN values and the duplicate values
# Continue to do some summary

citybike_combin.dtypes
```

```python
citybike_combin.describe()
```

```python
# To create a scatter plot of bike station locations
sns.scatterplot(x='longitude', y='latitude', data=citybike_combin,
                hue='num_bikes', palette='viridis')

# Add a legend and axis labels
plt.legend(title='Number of bikes')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Bike station locations and number of bikes')
plt.show()
```

```python
# To create a heatmap of the data
sns.heatmap(citybike_combin.corr(), cmap='coolwarm', annot=True)

plt.show()
```

```python
# To create a joint plot 
sns.jointplot(data=citybike_combin, x="restaurants", y="rating", hue="num_bikes")
plt.show()
```

```python
# To create pairplot for numbers of bikes with the number of restaurants
sns.pairplot(data=citybike_combin, hue="num_bikes")
plt.show()
```

# Database

Put all your results in an SQLite3 database (remember, SQLite stores its databases as files in your local machine - make sure to create your database in your project's data/ directory!)

```python
import sqlite3

# Create a new database file 
conn = sqlite3.connect('citybike.db')

# Write the citybike_combin DataFrame to a new table in the database
citybike_combin.to_sql('citybike_restaurants', conn, if_exists='replace', index=False)
```

Look at the data before and after the join to validate your data.

```python
# Connect to the database
conn = sqlite3.connect("citybike.db")

# Create a cursor
c = conn.cursor()

# Execute a query to retrieve all data from the table
c.execute("SELECT * FROM citybike_restaurants")

# Fetch all rows of the result
rows = c.fetchall()

# Print the data
for row in rows:
    print(row)

# Close the cursor and the connection
c.close()
conn.close()
```
